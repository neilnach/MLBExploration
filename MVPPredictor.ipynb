```python
%matplotlib inline

import pandas as pd
import numpy as np
from pylab import *
import seaborn as sns

#load tables
batting_reg = pd.DataFrame.from_csv('Batting.csv', index_col = None, encoding = 'utf-8')
batting_reg.tail(n=5)
```

```python
pitching_reg = pd.DataFrame.from_csv('Pitching.csv', index_col = None, encoding = 'utf-8')
pitching_reg.tail(n=5)
```

```python
player = pd.DataFrame.from_csv('Master.csv', index_col = None, encoding = 'utf-8')
player.tail(n=5)
```

```python
awards = pd.DataFrame.from_csv('AwardsPlayers.csv', index_col = None, encoding = 'utf-8')
awards.tail(n=5)
```

```python
player_name = player[['playerID', 'nameFirst', 'nameLast']]
#print (player_name)

#clean up csv for MVPs
mvp = awards[['playerID', 'awardID', 'yearID']]
#print (mvp[mvp['awardID'].str.contains('Most Valuable Player')])
temp_mvp = mvp.ix[(mvp['awardID'] == 'Most Valuable Player')]
temp_mvp.head(n=10)
```

```python
new_df = pd.merge(batting_reg, temp_mvp, on=['playerID', 'yearID'], how ='outer')
new_df.tail(n=5)
new_df['MVP'] = 0
new_df.MVP[new_df.awardID == 'Most Valuable Player'] = 1
new_df.loc[new_df['playerID'] == 'poseybu01']
```

```python
from sklearn.model_selection import train_test_split

new_df = new_df.fillna(0)

X = new_df.iloc[:, 5:22]  # select columns 1 through end
y = new_df.iloc[:, -1]   # select last column, the MVP status

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.7, random_state=7)

X_test.head()
```

```python
from sklearn.neighbors import KNeighborsClassifier

nbr = KNeighborsClassifier(n_neighbors=9)
nbr.fit(X_train, y_train)
nbr.score(X_test, y_test)
```

```python
from sklearn.decomposition import PCA
from sklearn.utils import resample

df_down = resample(new_df, n_samples=300)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.7, random_state=7)

nbr = KNeighborsClassifier(n_neighbors=9)
nbr.fit(X_train, y_train)
nbr.score(X_test, y_test)
```

```python
from matplotlib import pyplot as plt

plt.scatter(X_test[y_test == 0, 0], X_test[y_test == 0, 1], color = 'black')
plt.scatter(X_test[y_test == 1, 0], X_test[y_test == 1, 1], color = 'green')
plt.show()
```
